```python
def calc_sim(box1, box2):
    box1 = format_geom(box1)
    box2 = format_geom(box2)
    box1_ymin, box1_xmin, box1_ymax, box1_xmax = box1
    box2_ymin, box2_xmin, box2_ymax, box2_xmax = box2
    x_dist = min(abs(box1_xmin - box2_xmin), abs(box1_xmin - box2_xmax), abs(box1_xmax - box2_xmin), abs(
        box1_xmax - box2_xmax))
    y_dist = min(abs(box1_ymin - box2_ymin), abs(box1_ymin - box2_ymax), abs(box1_ymax - box2_ymin), abs(
        box1_ymax - box2_ymax))
    return (int(x_dist), int(y_dist))
```

```python
def format_geom(rect):
    tmp = np.array(rect)
    return [min(tmp[:, 1]), min(tmp[:, 0]), max(tmp[:, 1]), max(tmp[:, 0])]
```

```python
import fitz  # PyMuPDF
import pandas as pd
import re



# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        text += page.get_text()
    return text


# def parse_questions_and_answers(text):
#     # Define the regular expression patterns for parsing
#     pattern = re.compile(r'(\d+)\.\s+(.+?)(?=\n\(b\))', re.DOTALL)
#     sub_pattern = re.compile(r'\(b\)\s+(.+?)\n\s+(.+)', re.DOTALL)
    
#     matches = pattern.findall(text)
#     questions_and_answers = []

#     for match in matches:
#         parent_question = match[1].strip()
#         sub_match = sub_pattern.search(text[text.find(match[1]) + len(match[1]):])
#         if sub_match:
#             child_question = sub_match.group(1).strip()
#             child_answer = sub_match.group(2).strip()
#             questions_and_answers.append({
#                 "parent_question": parent_question,
#                 "child_question": child_question,
#                 "child_answer": child_answer
#             })
    
#     return questions_and_answers

def parse_questions_and_answers(text):
    questions_and_answers = []
    lines = text.split('\n')
    parent_answer = ""
    child_question = ""
    child_answer = ""
    in_child_question = False
    in_child_answer = False

    for i, line in enumerate(lines):
        stripped_line = line.strip()
        
        # Check for parent question indicator and skip it
        if re.match(r'^\d+\.', stripped_line):  # New parent question (we skip these)
            if parent_answer and child_question and child_answer:
                questions_and_answers.append({
                    "parent_answer": parent_answer,
                    "child_question": child_question,
                    "child_answer": child_answer
                })
            parent_answer = ""
            child_question = ""
            child_answer = ""
            in_child_question = False
            in_child_answer = False
        
        # Check for child question indicator
        elif re.match(r'^\(b\)', stripped_line):  # New child question
            in_child_question = True
            in_child_answer = False
            child_question = stripped_line[3:].strip()
        
        # Collect additional lines for the child question
        elif in_child_question and not stripped_line.startswith((' ', '\t')):  
            child_question += " " + stripped_line
        
        # Collect child answer based on indentation level
        elif in_child_question and (line.startswith('    ') or line.startswith('\t')):  # Child answer (indented)
            in_child_answer = True
            child_answer += " " + stripped_line
        
        # Collect additional lines for the parent answer
        elif not in_child_question:  # Collect parent answer
            parent_answer += " " + stripped_line

    if parent_answer and child_question and child_answer:
        questions_and_answers.append({
            "parent_answer": parent_answer,
            "child_question": child_question,
            "child_answer": child_answer
        })

    return questions_and_answers



# Path to the PDF file
pdf_path = '/Users/agah/Desktop/aaaa.pdf'

# Extract text from the PDF
text = extract_text_from_pdf(pdf_path)

# Parse questions and answers
questions_and_answers = parse_questions_and_answers(text)

# Format the data
formatted_data = format_questions_and_answers(questions_and_answers)

# Create a DataFrame
df = pd.DataFrame(formatted_data)

# Save the DataFrame to a CSV file
csv_path = '/Users/agah/Desktop/aaaa.csv'
df.to_csv(csv_path, index=False)
```
